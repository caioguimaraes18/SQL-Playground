```python
# 1. Qual a porcentagem de rotatividade de estoque por categoria?
# Mede a quantidade de estoque vendido ou utilizado em relação ao estoque inicial.

query = """
    select 
        categoria,
        round(((sum(cast(estoque_inicial as float)) - sum(cast(estoque_atual as float))) / sum(cast(estoque_inicial as float))) * 100,2) as rotatividade

  from estoque
  group by categoria
    """

df = pd.read_sql_query(query, conn)

df
```

```python
# 2. Qual a porcentagem de rotatividade de estoque por produto?
# Mede a quantidade de estoque vendido ou utilizado em relação ao estoque inicial.


query = """
    select 
        produto,
        round(((sum(cast(estoque_inicial as float)) - sum(cast(estoque_atual as float))) / sum(cast(estoque_inicial as float))) * 100,2) as rotatividade

  from estoque
  group by produto
    """

df = pd.read_sql_query(query, conn)

df
```

```python
# 3. Lead time médio de entrega por produto
# Mede o tempo médio entre o pedido e a entrega do produto.

query = """
    select  
            produto,
            julianday(data_ultima_saida) - julianday(data_reabastecimento)  as lead_time,
            julianday(data_proxima_entrega) - julianday(data_reabastecimento) as lead_time_proxima_entrega
            
  from estoque e 
  left join estoque_detalhado d on d.id = e.id
  group by produto
    """

df = pd.read_sql_query(query, conn)

df
```
